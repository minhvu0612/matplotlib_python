```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [10,34,5,23,14,19,16]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'b+--')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```

```python
plt.plot(x,y,'g<',alpha=0.5)
```
